```python
import pennylane as qml
from pennylane import numpy as np
import matplotlib.pyplot as plt
```

```python
dev = qml.device("default.qubit", wires=5)
```

```python
@qml.qnode(dev)
def circuit_1():

  for i in range(5):
    qml.Hadamard(wires=i)

  qml.CNOT(wires=[0, 1])
  qml.CNOT(wires=[1, 2])
  qml.CNOT(wires=[2, 3])
  qml.CNOT(wires=[3, 4])

  qml.SWAP(wires=[0, 4])

  qml.RX(np.pi/2, wires=2)

  return qml.state()
```

```python
# Execute circuit
circuit_1()
```

```python
# Draw the cricuit
fig, ax = qml.draw_mpl(circuit_1)()
plt.show()
```

```python
dev = qml.device("default.qubit", wires=4)

@qml.qnode(dev)
def circuit_2():

    # Apply a Hadamard gate to the first qubit (wire 0)
    qml.Hadamard(wires=0)

    # Rotate the second qubit (wire 1) by pi/3 around the X axis
    qml.RX(np.pi / 3, wires=1)

    # Apply Hadamard gates to the third and fourth qubits (wires 2 and 3)
    qml.Hadamard(wires=2)
    qml.Hadamard(wires=3)

    # Swap test part:
    # Apply Hadamard to the first and third qubits before swapping
    qml.Hadamard(wires=0)
    qml.Hadamard(wires=2)

    # Swap the states between qubits 0 and 2, and qubits 1 and 3 respectively
    qml.SWAP(wires=[0, 2])
    qml.SWAP(wires=[1, 3])

    # Final Hadamard gates to complete the swap test interference pattern
    qml.Hadamard(wires=0)
    qml.Hadamard(wires=2)

    return qml.state()

# Execute and print the state vector
final_state = circuit_2()
print("Final state vector:", final_state)

# Draw the circuit
fig, ax = qml.draw_mpl(circuit_2)()
plt.show()
```
